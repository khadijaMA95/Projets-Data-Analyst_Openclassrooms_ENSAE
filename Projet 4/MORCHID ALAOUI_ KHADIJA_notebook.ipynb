**Pour calculer le résultat de cette question, nous suivons les étapes suivantes :**

```python
#Calcul du nombre d'humains pouvant être nourris

kcal_total = dispo_alimentaire_2017["dispo_kcal_pays"].sum()
kcal_moyenne = 2500
pers_nourrie = round((kcal_total/kcal_moyenne))

print("La disponibilité mondiale en nourriture en 2017 était de",kcal_total, "Ainsi le nombre de personne pouvant être nourries théoriquement en 2017 sur une base de 2500 Kcal par jour  était",pers_nourrie)
```

```python
# calcul du proportion de la population totale qui pourrait être nourries par la disponibilité totale

pop_totale = dispo_alimentaire_2017["Population"].sum()
proportion_personne_nourrie = round((pers_nourrie / pop_totale) * 100,1)
print(" La proportion de la population totale qui pourrait être nourrie par la disponibilité totale représente", proportion_personne_nourrie,"% de la population mondiale totale . Donc on peut dire que l'alimentation disponible n'est pas équitablement répartie dans le monde !")
```
